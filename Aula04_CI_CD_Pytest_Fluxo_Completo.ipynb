# üîÑ Passo 4 ‚Äì Versionando com Git


````python
git add .
git commit -m "test: adiciona teste da rota home"
git push origin main
````

git add ‚Üí prepara altera√ß√µes

# ü§ñ Passo 5 ‚Äì Integrando no GitHub Actions

Arquivo: .github/workflows/ci.yml


````python
name: CI - API Flask

on:
  push:
    branches:
      - main

jobs:
  build:
    runs-on: ubuntu-latest

    steps:
      - uses: actions/checkout@v3

      - uses: actions/setup-python@v4
        with:
          python-version: '3.11'

      - run: pip install -r requirements.txt

      - run: pytest
````

Sempre que ocorrer push na main:

- Runner √© criado
- C√≥digo √© baixado
- Depend√™ncias s√£o instaladas
- pytest √© executado
